```python


# 1) ¿Cómo varía el coste médico anual según edad, sexo y región?
if {"age_bin","sex","annual_medical_cost"}.issubset(df.columns):
    data1 = df.groupby(["age_bin","sex"])["annual_medical_cost"].mean().reset_index()
    plt.figure(figsize=(10,5))
    sns.lineplot(data=data1, x="age_bin", y="annual_medical_cost", hue="sex", marker="o")
    plt.title("Coste medio anual por grupo de edad y sexo")
    plt.xlabel("Grupo de edad")
    plt.ylabel("Annual medical cost (mean)")
    plt.show()

if {"region","sex","annual_medical_cost"}.issubset(df.columns):
    pivot = df.groupby(["region","sex"])["annual_medical_cost"].mean().unstack()
    plt.figure(figsize=(10,6))
    sns.heatmap(pivot, annot=True, fmt=".0f", cmap="viridis")
    plt.title("Coste medio anual por región y sexo")
    plt.xlabel("Sexo")
    plt.ylabel("Región")
    plt.show()
```

```python


# 2) ¿Qué relación existe entre hábitos de vida (bmi, smoker, alcohol_freq…) y el coste médico anual?
# 2a) BMI vs cost (scatter + lowess-like smoothing using rolling median by BMI bucket)
if {"bmi","annual_medical_cost"}.issubset(df.columns):
    samp = df.sample(min(10000, len(df)), random_state=1)
    plt.figure(figsize=(9,5))
    sns.scatterplot(x="bmi", y="annual_medical_cost", data=samp, alpha=0.15, s=8)
    # smooth: groupby bmi rounded
    tmp = samp.assign(bmi_round = samp["bmi"].round(0)).groupby("bmi_round")["annual_medical_cost"].median().reset_index()
    sns.lineplot(x="bmi_round", y="annual_medical_cost", data=tmp, color="red")
    plt.title("BMI vs Annual medical cost (muestra) + tendencia")
    plt.show()

# 2b) Smoker vs cost (boxplot)
if {"smoker","annual_medical_cost"}.issubset(df.columns):
    # asegurarse de que smoker tenga valores 0/1 o categorías
    plt.figure(figsize=(7,4))
    sns.boxplot(x="smoker", y="annual_medical_cost", data=df)
    plt.yscale("symlog")
    plt.title("Distribución coste por estado de fumador")
    plt.xlabel("Smoker")
    plt.show()
```

Aquí empezamos a investigar aquellos factores más considerados de riesgo, entre ellos el del tabaco. Se puede observar que, bien fumes ahora mismo o lo hayas hecho en el pasado, eres más propenso a sufrir complicaciones que, para la compañía, se traduzcan en gastos. Así pues, se podría empezar a sugerir que aquellos clientes fumadores van a suponer mayor coste para la compañía y se deberían intentar evitar o crear una tarifa especial para ellos. 

```python



# 3) ¿Cuánto más cuesta un paciente de alto riesgo respecto a uno de bajo riesgo?
if {"is_high_risk","annual_medical_cost"}.issubset(df.columns):
    agg = df.groupby("is_high_risk")["annual_medical_cost"].agg(["mean","median","count"]).reset_index()
    # barplot means
    plt.figure(figsize=(6,4))
    sns.barplot(x="is_high_risk", y="mean", data=agg.rename(columns={"mean":"mean"}))
    plt.title("Coste medio anual: alto riesgo vs bajo riesgo")
    plt.ylabel("Annual medical cost (mean)")
    plt.xlabel("is_high_risk")
    plt.show()
    # boxplot to show distribution
    plt.figure(figsize=(6,4))
    sns.boxplot(x="is_high_risk", y="annual_medical_cost", data=df)
    plt.yscale("symlog")
    plt.title("Distribución de costes por riesgo")
    plt.show()
```

Las dos figuras permiten conocer un poco más a fondo la propia dicotomía que ofrece la compañía. Divide a sus pacientes entre aquellos de riesgo y aquellos 'sanos'. Se puede ver que no es una elección hecha a voleo: aquellos considerados como sanos cuestan casi la mitad que los enfermos. Más adelante nos encontraremos con risk_score y con varias variables cuantitativas que indaguen un poco más y nos den pistas sobre en base a qué se ha tomado esta decisión.

```python
    
chronic_cols = [c for c in ["diabetes","hypertension","cardiovascular","copd","kidney_disease","liver_disease","arthritis","cancer_history"] if c in df.columns]
if len(chronic_cols) > 0:
    means = []
    for c in chronic_cols:
        m = df.loc[df[c]==1, "annual_medical_cost"].mean()
        means.append((c,m, df[c].sum()))
    means_df = pd.DataFrame(means, columns=["condition","mean_cost","count"]).sort_values("mean_cost", ascending=False)
    plt.figure(figsize=(8,5))
    sns.barplot(x="mean_cost", y="condition", data=means_df)
    plt.title("Coste medio anual por enfermedad crónica (solo pacientes con la condición)")
    plt.xlabel("Annual medical cost (mean)")
    plt.show()
plt.subplot(3, 2, 5)
```

```python
# 5) ¿Existe relación entre hospitalizaciones y coste total en los últimos 3 años?
if {"hospitalizations_last_3yrs","annual_medical_cost","total_claims_paid"}.issubset(df.columns):
    plt.figure(figsize=(8,5))
    sns.scatterplot(x="hospitalizations_last_3yrs", y="annual_medical_cost", data=df.sample(min(5000,len(df))), alpha=0.5)
    sns.regplot(x="hospitalizations_last_3yrs", y="annual_medical_cost", data=df.sample(min(2000,len(df))), scatter=False, color="red")
    plt.title("Hospitalizaciones (3 yrs) vs Annual medical cost")
    plt.show()

    plt.figure(figsize=(8,5))
    sns.scatterplot(x="hospitalizations_last_3yrs", y="total_claims_paid", data=df.sample(min(5000,len(df))), alpha=0.5)
    sns.regplot(x="hospitalizations_last_3yrs", y="total_claims_paid", data=df.sample(min(2000,len(df))), scatter=False, color="red")
    plt.title("Hospitalizaciones (3 yrs) vs Total claims paid")
    plt.show()
```

```python


# 6) ¿Qué parámetros clínicos tienen mayor correlación con el coste médico?
clinical_feats = [c for c in ["age","bmi","chronic_count","systolic_bp","diastolic_bp","ldl","hba1c","risk_score","visits_last_year","medication_count"] if c in df.columns]
if len(clinical_feats) >= 2 and "annual_medical_cost" in df.columns:
    corr_df = df[clinical_feats + ["annual_medical_cost"]].corr()
    plt.figure(figsize=(10,8))
    sns.heatmap(corr_df, annot=True, fmt=".2f", cmap="vlag")
    plt.title("Correlación entre parámetros clínicos y coste médico")
    plt.show()
    # Order features by absolute correlation with cost
    corr_with_cost = corr_df["annual_medical_cost"].abs().drop("annual_medical_cost").sort_values(ascending=False)
    plt.figure(figsize=(6,4))
    sns.barplot(x=corr_with_cost.values, y=corr_with_cost.index)
    plt.title("Correlación absoluta con annual_medical_cost")
    plt.xlabel("abs(correlation)")
    plt.show()
```

Aquí podemos ver que existe una variable llamada risk_score que es la más correlacionada con el coste. Por lo tanto, podemos deducir que esta variable ha sido extraida de un compendio de todas las demás variables negativas (aún sin saber con qué peso cada una)

```python

# 8) ¿Cómo se relacionan los procedimientos médicos con el total pagado en reclamaciones?
proc_cols = [c for c in ["proc_imaging","proc_surgery","proc_psycho","proc_consult_count","proc_lab"] if c in df.columns]
if len(proc_cols) > 0 and "total_claims_paid" in df.columns:
    # Scatter cada procedimiento vs total claims (muestra)
    sample = df.sample(min(5000, len(df)), random_state=1)
    for c in proc_cols:
        plt.figure(figsize=(8,4))
        sns.scatterplot(x=c, y="total_claims_paid", data=sample, alpha=0.4)
        sns.regplot(x=c, y="total_claims_paid", data=sample.sample(min(2000,len(sample))), scatter=False, color="red")
        plt.title(f"{c} vs total_claims_paid")
        plt.show()
    # También un heatmap de correlaciones entre procedimientos y total_claims_paid
    tmp = df[proc_cols + ["total_claims_paid"]].corr()
    plt.figure(figsize=(8,6))
    sns.heatmap(tmp, annot=True, fmt=".2f", cmap="vlag")
    plt.title("Correlación entre procedimientos y total_claims_paid")
    plt.show()
```

```python

# 10) ¿Qué grupos de pacientes comparten patrones similares de riesgo y coste (clustering)?

cluster_features = [c for c in ["age","bmi","chronic_count","risk_score","annual_medical_cost","visits_last_year","hospitalizations_last_3yrs","medication_count"] if c in df.columns]
if len(cluster_features) >= 3:
    Xc = df[cluster_features].fillna(0)
    scaler = StandardScaler()
    Xc_s = scaler.fit_transform(Xc)

    pca = PCA(n_components=2, random_state=42)
    Xp = pca.fit_transform(Xc_s)

    kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
    labs = kmeans.fit_predict(Xc_s)
    df["_cluster"] = labs

    # scatter PCA colored by cluster
    plt.figure(figsize=(8,5))
    sns.scatterplot(x=Xp[:,0], y=Xp[:,1], hue=labs, palette="tab10", s=20)
    plt.title("Clusters (PCA 2D) — 4 clusters")
    plt.xlabel("PC1")
    plt.ylabel("PC2")
    plt.legend(title="cluster")
    plt.show()

    # perfil de clusters: cost & risk
    prof = df.groupby("_cluster")[["risk_score"] + [c for c in cluster_features if c not in ["annual_medical_cost","risk_score"]]].median()
    prof = prof.reset_index().melt(id_vars="_cluster", var_name="feature", value_name="median")
    plt.figure(figsize=(10,6))
    sns.barplot(x="feature", y="median", hue="_cluster", data=prof)
    plt.title("Perfil mediano por cluster (coste, riesgo y otras métricas)")
    plt.xticks(rotation=45)
    plt.show()
```

```python
# 10) ¿Qué grupos de pacientes comparten patrones similares de riesgo y coste (clustering)?

cluster_features = [c for c in ["age","bmi","chronic_count","risk_score","annual_medical_cost","visits_last_year","hospitalizations_last_3yrs","medication_count"] if c in df.columns]
if len(cluster_features) >= 3:
    Xc = df[cluster_features].fillna(0)
    scaler = StandardScaler()
    Xc_s = scaler.fit_transform(Xc)

    pca = PCA(n_components=2, random_state=42)
    Xp = pca.fit_transform(Xc_s)

    kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
    labs = kmeans.fit_predict(Xc_s)
    df["_cluster"] = labs

    # scatter PCA colored by cluster
    plt.figure(figsize=(8,5))
    sns.scatterplot(x=Xp[:,0], y=Xp[:,1], hue=labs, palette="tab10", s=20)
    plt.title("Clusters (PCA 2D) — 4 clusters")
    plt.xlabel("PC1")
    plt.ylabel("PC2")
    plt.legend(title="cluster")
    plt.show()

    # perfil de clusters: cost & risk
    prof = df.groupby("_cluster")[["risk_score"] + [c for c in cluster_features if c not in ["annual_medical_cost","risk_score"]]].median()
    prof = prof.reset_index().melt(id_vars="_cluster", var_name="feature", value_name="median")
    plt.figure(figsize=(10,6))
    sns.barplot(x="feature", y="median", hue="_cluster", data=prof)
    plt.title("Perfil mediano por cluster (coste, riesgo y otras métricas)")
    plt.xticks(rotation=45)
    plt.show()
```
